### Another thing we can do to prevent overfitting is adding a dropout layer (it will turn off a percentage of neurons randomly)

```python
from tensorflow.keras.layers import Dropout
```

```python
model = Sequential()

model.add(Dense(30, activation='relu'))
#after the two dense layers we add a dropout layer 
#as seen the dropout takes an argument (rate: the percentage of neurons to turn off)
model.add(Dropout(0.5))
model.add(Dense(15, activation='relu'))
model.add(Dropout(0.5))
#binary calssification
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
model.fit(x=X_train, y=y_train, epochs=600, validation_data=(X_test, y_test), 
         callbacks=[early_stop])
```

```python
model_loss=pd.DataFrame(model.history.history)
```

```python
model_loss.plot()
```

A much better behavior as the training and validation loss both goes down quickly

```python
predictions = model.predict(X_test)

predicted_classes = (predictions > 0.5).astype(int)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test, predicted_classes))
```

```python
print(classification_report(y_test, predicted_classes))
```
